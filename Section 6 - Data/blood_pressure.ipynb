```python
import pandas as pd
```

```python
df = pd.read_csv('data2.csv')
df
```

```python
def get_column_name(column):
    if column == 'Unnamed: 0':
        return 'Country'
    if len(column) == 4:
        return column + ' (Both sexes)'
    elif column[-1] == '1':
        return column[0:4] + ' (Male)'
    elif column[-1] == '2':
        return column[0:4] + ' (Female)'
```

```python
columns = [get_column_name(column) for column in df.columns]
```

```python
df.columns = columns
```

```python
df = df.iloc[3:, 0:]
df
```

```python
def remove_range(cell):
    return cell.split(' ')[0]
```

```python
for column in df.columns:
    if column != 'Country':
        df[column] = df[column].apply(remove_range)
df
```
