# THE DATASET
This is the dataset downloaded from the WHO website. As you can see, it's not pretty...

```python
df = pd.read_csv('data.csv')
df.head()
```

# Renaming columns
Let's start by renaming the columns (these are in bold at the top of the dataframe above). We will create a function which takes the old column name as input and returns a new improved column name.

```python
def get_column_name(column):
    if column.name == 'Unnamed: 0':
        return 'Country'
    else:
        return column.name[:4] + f' ({column.iloc[2][1:]})'
```

Now let's assign the columns property of the dataframe (df.columns) to a list where each element of the list is generated by insterting 

```python
df.columns = [get_column_name(df[column]) for column in df.columns]
```

# Removing the 1st 3 rows

```python
df = df.iloc[3:, 0:]
df
```

# Making the country column the index of df

```python
df = df.set_index('Country')

df
```

# Removing the range part of the blood pressure data

```python
def remove_range(element):
    return element.split(' ')[0]
```

```python
for column in df.columns:
    if column != 'Country':
        df[column] = df[column].apply(remove_range)
df
```

## Creating a new dataset (df_both) which only contains the (BothSexes) Columns from df

```python
lst = []

for column in df.columns[1:]:
    print(column.split(' ')[1])
    if column.split(' ')[1] != '(Both ':
        lst.append(column)

df_both = df.drop(labels=lst, axis=1)

df_both.columns = [column.split(' ')[0] for column in df_both.columns]
```

# Realising that for some countries, there is "No" data:

```python
df_both.loc['Sudan']
```

# Finding these countries

```python
countries = df_both.index

countries_to_remove = []

for country in countries:
    if df_both.loc[country]['2015'] == 'No':
        countries_to_remove.append(country)
        
countries_to_remove
```

# Removing these countries

```python
df_both = df_both.drop(countries_to_remove)
```

# Converting the string values to floats

```python
for column in df_both.columns:
    df_both[column] = pd.to_numeric(df_both[column], downcast='float')
```

### Transposing the data so that the years become the rows to fit with seaborn lineplot() function

```python
df_both = df_both.transpose()

df_both
```

## Creating a line plotting function which takes countries to plot as arguments

```python
def line_plot(countries):
    plt.figure(figsize=(10,6))

    lst = np.arange(0,41,5)
    plt.xticks(lst)
    
    plt.title("How has blood pressure has varied in different countries between 1975 and 2015")
    plt.ylabel('Blood pressure')
    plt.xlabel('Year')

    sns.lineplot(data=df_both[countries].iloc[::-1])
    
line_plot(['Uruguay', 'United States of America', 'Zambia', 'Zimbabwe'])
```
